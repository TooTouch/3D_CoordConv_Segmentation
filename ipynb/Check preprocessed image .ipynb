```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

import h5py
```

# Load data

```python
train_set = h5py.File('../dataset/h5py/train_hf128_7','r')
```

```python
list(train_set.keys())
```

# CT image

```python
ct_image = np.array(train_set['ct_image_0'])
ct_image_mask = np.array(train_set['ct_image_mask_0'])
ct_label = np.array(train_set['ct_label_0'])

print('ct_image shape: ',ct_image.shape)
print('ct_image_mask shape: ',ct_image_mask.shape)
print('ct_label shape: ',ct_label.shape)
```

```python
ct_image = ct_image.reshape(ct_image.shape[:-1])
ct_image_mask = ct_image_mask.reshape(ct_image_mask.shape[:-1])
ct_label = np.argmax(ct_label, axis=-1)

print('ct_image shape: ',ct_image.shape)
print('ct_image_mask shape: ',ct_image_mask.shape)
print('ct_label shape: ',ct_label.shape)
```

```python
f, ax = plt.subplots(1,3, figsize=(10,5))
ax[0].imshow(ct_image[64,:,:])
ax[1].imshow(ct_image_mask[64,:,:])
ax[2].imshow(ct_label[64,:,:])
plt.show()
f, ax = plt.subplots(1,3, figsize=(10,5))
ax[0].imshow(ct_image[:,64,:])
ax[1].imshow(ct_image_mask[:,64,:])
ax[2].imshow(ct_label[:,64,:])
plt.show()
f, ax = plt.subplots(1,3, figsize=(10,5))
ax[0].imshow(ct_image[:,:,64])
ax[1].imshow(ct_image_mask[:,:,64])
ax[2].imshow(ct_label[:,:,64])
plt.show()
```
